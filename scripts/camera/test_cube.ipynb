```python
import sys
sys.path.append("../homography")
sys.path.append("../camera")
sys.path.append("../local_image_descriptors")
import homography
import camera
import sift
import cube
%pylab inline

# compute features                                                        
#sift.process_image('../../data/book_frontal.JPG','../../data/im0.sift')
im1='../../data/space_front.jpg'
im2='../../data/space_perspective.jpg'
im1='../../data/mag_front.jpg'
#im2='../../data/mag_perspective.jpg'
im2='../../data/mag_perspective_1.jpg'
ims = [im1, im2]
#im1='../../data/book_frontal.JPG'
#im2='../../data/book_perspective.JPG'
#im1='../../data/hp_frontal.jpg'
#im2='../../data/hp_perspective.jpg'
sifts = []
for k in range(2):
    sifts.append(ims[k][:-4]+".sift")
    #sift.process_image(ims[k], sifts[k])

l0,d0 = sift.read_features_from_file(sifts[0])
#sift.process_image('../../data/book_perspective.JPG','../../data/im1.sift')
l1,d1 = sift.read_features_from_file(sifts[1])
# match features and estimate homography                                        
#matches = sift.match(d0,d1)
matches = sift.match_twosided(d0,d1)
ndx = matches.nonzero()[0]
fp = homography.make_homog(l0[ndx,:2].T)
ndx2 = [int(matches[i]) for i in ndx]
tp = homography.make_homog(l1[ndx2,:2].T)
model = homography.RansacModel()
```

```python
from PIL import Image
print len(ndx2)
img1 = array(Image.open(im1))
img2 = array(Image.open(im2))


#img1 = Image.fromarray(img1).transpose(Image.ROTATE_90)
img1 = Image.fromarray(img1)
img2 = Image.fromarray(img2)
imshow(vstack((img1, img2)))
#print img1.size
#print img2.size
#print l0
for i,x in enumerate(ndx):
    #plot(l0[x][1],l0[x][0],'r+')
    #plot(l1[ndx2[i]][0],l1[ndx2[i]][1]+img1.size[1],'r+')
    #xs = [l0[x][1], l1[ndx2[i]][0]]
    xs = [l0[x][0], l1[ndx2[i]][0]]
    ys = [l0[x][1], l1[ndx2[i]][1]+img1.size[1]]
    plot(xs, ys, lw=1, color="red")
```

```python
H,ransac_data = homography.H_from_ransac(fp,tp,model)
# camera calibration
#K = camera.my_calibration((747,1000))
K = camera.my_calibration((2048,1152))
# 3D points at plane z=0 with sides of length 0.2
box = cube.cube_points([0,0,0.1],0.1)
# project bottom square in first image
cam1 = camera.Camera( hstack((K,dot(K,array([[0],[0],[-1]])) )) )
# first points are the bottom square
box_cam1 = cam1.project(homography.make_homog(box[:,:5]))
# use H to transfer points to the second image

#imshow(vstack((img1, img2)))
imshow(img1)
pts = array([[600,600,2000,2000], [50,1100,1100,50], [1,1,1,1]])
plot(pts[0], pts[1], lw=3, color="blue")
pts2 = dot(H,pts)
figure()
imshow(img2)
plot(pts2[0]/pts2[2], pts2[1]/pts2[2], lw=3, color="green")

""" It seems that even as few as 38 matching SIFT points were enough to estimate the homography well """
```
